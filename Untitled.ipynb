```python
# Type of tagert encoder :
def calc_smooth_mean(df, by, on, m):
    # Compute the global mean
    mean = df[on].mean()

    # Compute the number of values and the mean of each group
    agg = df.groupby(by)[on].agg(['count', 'mean'])
    counts = agg['count']
    means = agg['mean']

    # Compute the "smoothed" means
    smooth = (counts * means + m * mean) / (counts + m)

    # Replace each value by the according smoothed mean
    return df[by].map(smooth)
```

```python
trainx['Surroundings'] = calc_smooth_mean(property_data, by='Surroundings', on='PropertyEvaluationvalue', m=300)/trainx["TotalNoOfUnits"]
trainx['Block'] = calc_smooth_mean(property_data, by='Block', on='PropertyEvaluationvalue', m=300)/trainx["TotalNoOfUnits"]
trainx['ZipCode'] = calc_smooth_mean(property_data, by='ZipCode', on='PropertyEvaluationvalue', m=300)/trainx["TotalNoOfUnits"]


validationx['Surroundings'] = calc_smooth_mean(property_data, by='Surroundings', on='PropertyEvaluationvalue', m=300)/validationx["TotalNoOfUnits"]
validationx['Block'] = calc_smooth_mean(property_data, by='Block', on='PropertyEvaluationvalue', m=300)/validationx["TotalNoOfUnits"]
validationx['ZipCode'] = calc_smooth_mean(property_data, by='ZipCode', on='PropertyEvaluationvalue', m=300)/validationx["TotalNoOfUnits"]


test_data['Surroundings'] = calc_smooth_mean(property_data, by='Surroundings', on='PropertyEvaluationvalue', m=300)/test_data["TotalNoOfUnits"]
test_data['Block'] = calc_smooth_mean(property_data, by='Block', on='PropertyEvaluationvalue', m=300)/test_data["TotalNoOfUnits"]
test_data['ZipCode'] = calc_smooth_mean(property_data, by='ZipCode', on='PropertyEvaluationvalue', m=300)/test_data["TotalNoOfUnits"]



```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
reg = LinearRegression()
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
#Fit the data(train the model)
model= reg.fit(X_train_aligned, trainy)
# Predict
y_pred_train = model.predict(X_train_aligned)
y_pred_test = model.predict(X_validation_aligned)
# model evaluation

mse_train = mean_squared_error(trainy, y_pred_train)
mse_test = mean_squared_error(validationy, y_pred_test)

rmse_train = np.sqrt(mean_squared_error(trainy, y_pred_train))
rmse_test = np.sqrt(mean_squared_error(validationy, y_pred_test))

mae_train = mean_absolute_error(trainy, y_pred_train)
mae_test = mean_absolute_error(validationy, y_pred_test)

r2 =model.score(X_train_aligned, trainy)
# printing values
print('Slope:' ,model.coef_)
print('Intercept:', model.intercept_)
print('mean squared error for train: ', mse_train)
print('mean squared error for validation: ', mse_test)
print('root mean squared error for train: ', rmse_train)
print('root mean squared error validation: ', rmse_test)
print('mean absolute error for train: ', mae_train)
print('mean absolute error for validation: ', mae_test)
print('mean absolute percentage error for train: ', mean_absolute_percentage_error(trainy, y_pred_train))
print('mean absolute percentage error for validation: ', mean_absolute_percentage_error(validationy, y_pred_test))

print('R2 score: ', r2)
```

```python
from mlxtend.plotting import plot_learning_curves
```

```python
# Function for error metrics for non - linear model
def regression_error_metrics(trainy, trainy_pred, testy, testy_pred):

    mse_train = mean_squared_error(trainy, trainy_pred)
    mse_test = mean_squared_error(testy, testy_pred)

    rmse_train = np.sqrt(mean_squared_error(trainy, trainy_pred))
    rmse_test = np.sqrt(mean_squared_error(testy, testy_pred))

    mae_train = mean_absolute_error(trainy, trainy_pred)
    mae_test = mean_absolute_error(testy, testy_pred)

    # printing values


    print('mean squared error for train: ', mse_train)
    print('mean squared error for validation: ', mse_test)
    print('root mean squared error for train: ', rmse_train)
    print('root mean squared error validation: ', rmse_test)
    print('mean absolute error for train: ', mae_train)
    print('mean absolute error for validation: ', mae_test)
    print('mean absolute percentage error for train: ', mean_absolute_percentage_error(trainy, trainy_pred))
    print('mean absolute percentage error for validation: ', mean_absolute_percentage_error(testy, testy_pred))

```
